```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Exploracion de datos

```python
data = pd.read_csv('Cancer_de_pulmones.csv')
print(data.head())
print(data.info())
```

#### Estadistica descriptiva

```python
print(data.describe())
```

#### Revicion de faltantes

```python
print(data.isnull().sum())
```

#### Variable Categorica a predecir (LUNG_CANCER)

```python
print(data['LUNG_CANCER'].value_counts())
```

```python
sns.countplot(x='LUNG_CANCER', data=data)
plt.title('Distribuci√≥n de LUNG_CANCER')
plt.xlabel('LUNG_CANCER')
plt.ylabel('Frecuencia')
plt.show()
```

#### Variables predictoras

```python
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=data)
plt.title('Relaci√≥n entre Fumar y LUNG_CANCER')
plt.xlabel('Fumar')
plt.ylabel('Frecuencia')
plt.legend(title='LUNG_CANCER')
plt.show()
```

```python

sns.countplot(x='GENDER', hue='LUNG_CANCER', data=data)
plt.title('Distribuci√≥n de LUNG_CANCER por G√©nero')
plt.xlabel('G√©nero')
plt.ylabel('Frecuencia')
plt.legend(title='LUNG_CANCER')
plt.show()
```

```python
sns.lmplot(x='AGE', y='SMOKING', hue='LUNG_CANCER', data=data, aspect=1.5, ci=None)
plt.title('Relaci√≥n entre Edad y Fumado con respecto a LUNG_CANCER')
plt.xlabel('Edad')
plt.ylabel('Fumado')
plt.show()
```

```python
sns.lmplot(x='AGE', y='ALCOHOL_CONSUMING', hue='LUNG_CANCER', data=data, aspect=1.5, ci=None)
plt.title('Relaci√≥n entre Edad y Consumo de Alcohol con respecto a LUNG_CANCER')
plt.xlabel('Edad')
plt.ylabel('Consumo de Alcohol')
plt.show()
```

```python
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=data)
plt.title('Distribuci√≥n de LUNG_CANCER por Fumado')
plt.xlabel('Fumado')
plt.ylabel('Frecuencia')
plt.legend(title='LUNG_CANCER')
plt.show()
```

```python
sns.countplot(x='YELLOW_FINGERS', hue='LUNG_CANCER', data=data)
plt.title('Distribuci√≥n de LUNG_CANCER por Dedos Amarillos')
plt.xlabel('Dedos Amarillos')
plt.ylabel('Frecuencia')
plt.legend(title='LUNG_CANCER')
plt.show()
```

```python
sns.countplot(x='CHEST_PAIN', hue='LUNG_CANCER', data=data)
plt.title('Distribuci√≥n de LUNG_CANCER por Dolor de Pecho')
plt.xlabel('Dolor de Pecho')
plt.ylabel('Frecuencia')
plt.legend(title='LUNG_CANCER')
plt.show()
```
